```python
mnt_total = data['MntTotal']
iqr = mnt_total.quantile(0.75) - mnt_total.quantile(0.25)
lower = mnt_total.quantile(0.25) - 1.5 * iqr
upper = mnt_total.quantile(0.75) + 1.5 * iqr
outliers = data[(mnt_total < lower) | (mnt_total > upper)]
outliers.head()
```

```python
data=data[(data['MntTotal']>lower)& (data['MntTotal']<upper)]
```

```python
def get_marital_status(row):
  if row['marital_Divorced'] == 1:
    return 'Divorced'
  elif row['marital_Married'] == 1:
    return 'Married'
  elif row['marital_Single'] == 1:
    return 'Single'
  elif row['marital_Together'] == 1:
    return 'Together'
  elif row['marital_Widow'] == 1:
    return 'Widow'
  else:
    return 'Unknown'
data['Marital']= data.apply(get_marital_status, axis=1)
```

```python
sns.barplot(x='Marital', y='MntTotal', data=data)
plt.title('Marital Status vs. MntTotal')
plt.xlabel('Marital Status')
plt.ylabel('MntTotal')
```
